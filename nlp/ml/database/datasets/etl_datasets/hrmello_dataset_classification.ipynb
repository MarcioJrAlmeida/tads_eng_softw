### Gerando Perguntas para o Contexto

```python
def gerar_contexto(resposta):
    texto = str(resposta).lower()

    if 'gorda' in texto or 'obesa' in texto or 'peso' in texto:
        return 'Qual sua opinião sobre o corpo das pessoas?'
    
    elif 'muçulman' in texto:
        return 'O que você pensa sobre a religião islâmica ou seus seguidores?'
    
    elif 'judeu' in texto or 'judaico' in texto:
        return 'Como você enxerga o povo judeu e sua cultura?'
    
    elif 'mulher' in texto:
        return 'Qual sua opinião sobre o papel das mulheres na sociedade?'
    
    elif 'homem' in texto:
        return 'O que você pensa sobre o comportamento masculino?'
    
    elif 'nordestin' in texto:
        return 'Qual sua visão sobre os nordestinos e sua cultura?'
    
    elif 'preto' in texto or 'negro' in texto:
        return 'O que você pensa sobre as pessoas negras?'
    
    elif 'lgbt' in texto or 'gay' in texto or 'lésbica' in texto or 'trans' in texto:
        return 'Qual sua opinião sobre a comunidade LGBTQIAP+?'
    
    elif 'imigrante' in texto or 'estrangeiro' in texto:
        return 'Como você vê os imigrantes no Brasil ou no mundo?'
    
    elif 'idoso' in texto or 'velho' in texto:
        return 'Qual sua opinião sobre as pessoas idosas?'
    
    elif 'pobre' in texto:
        return 'Como você enxerga a pobreza e as pessoas em vulnerabilidade social?'
    
    else:
        return 'Qual sua opinião sobre esse assunto?'
```

```python
df_transformado_hate['contexto_pergunta'] = df_transformado_hate['conteudo_resposta'].apply(gerar_contexto)
```

```python
df_transformado_hate.head(5)
```

```python
## Ordenar as colunas
ordem_colunas = [
    'contexto_pergunta', 
    'conteudo_resposta',
    'classificada_como_ofensiva',
    'classificada_como_sentimento',
    'observacao',
    'original_idx',
    'valor'
]

df_transformado_hate = df_transformado_hate[ordem_colunas]
df_transformado_hate.drop(['original_idx', 'valor'], axis=1, inplace=True)
```

### Gerando Inserts para o Banco

```python
# 1️⃣ Conexão com SQL Server
conn = get_connection_ml()
cursor = conn.cursor()

# 2️⃣ Montagem das linhas
linhas_insert = []

for index, row in df_transformado_hate.iterrows():
    contexto = str(row['contexto_pergunta']).replace("'", "''")
    resposta = str(row['conteudo_resposta']).replace("'", "''")
    ofensiva = int(row['classificada_como_ofensiva'])
    sentimento = str(row['classificada_como_sentimento']).replace("'", "''")
    observacao = str(row['observacao']).replace("'", "''")

    linha = f"('{contexto}', '{resposta}', {ofensiva}, '{sentimento}', '{observacao}')"
    linhas_insert.append(linha)

# 3️⃣ Quebrar em blocos de até 1000 e executar
blocos = [linhas_insert[i:i+1000] for i in range(0, len(linhas_insert), 1000)]

for bloco in blocos:
    valores_sql = ",\n".join(bloco)
    comando_full = f"""INSERT INTO Treinamento_Manual (
        contexto_pergunta, conteudo_resposta, 
        classificada_como_ofensiva, classificada_como_sentimento, observacao
    ) VALUES
    {valores_sql};"""

    cursor.execute(comando_full)
    conn.commit()  # Aplica as inserções no banco

# 4️⃣ Encerrar conexão
cursor.close()
conn.close()
```
